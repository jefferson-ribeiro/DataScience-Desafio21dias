```
import pandas as pd

data = {
        'nome':['Narciso','Isa', 'Liah', 'Lana'],
        'idade': [33, 28, 0 , 6],
        'cidade': ['SP', 'MG', 'ABC', 'ABC']
}

df = pd.DataFrame(data)

# print(df)
display(df)
```

```
print(type(df)) # o tipo de variavel df
```

```
display( df.loc[0] ) # retorna o primeiro indice da lista
```

```
item = df.loc[0]
print(type(item)) # tipo do objeto de dentro de um dataframe 
```

```
df.loc[:,['nome', 'cidade']] #seleciona todas as linhas das ccolunas listadas
```

```
df.loc[0:2,['nome', 'idade']] # retorna da linha 0 até a linha 2 de um dataframe
```

```
df.loc[0:2, 'nome'] # retorna da linha 0 até a linha 2 mostrando somente uma coluna
```

```
print(type(df.loc[0:2, 'nome']))
```


<class 'pandas.core.series.Series'>  - Series: Representa uma linha ou uma coluna no dataframe



```
df.loc[[0,2]] # retorna linhas pelo indice da linha no caso 0 = linha 1 e 2 = linha 3
```

```
df.loc[(df['idade'] >= 20) & (df['idade'] <= 40)] # busccando por coluna entre valores
```

```
df.loc[0:2] # retorna da linha 0 ate a linha dois mostrando todas a colunas
```

```
df.iloc[0,1] # seleciona a primeira linha e segunda coluna
```

```
df.iloc[:,[1,2]] # seleciona as colunas do indice 1 e 2 com todas as linhas
```

```
df.iloc[[0,2],:] # seleciona linha 0 e 2 com todas as colunass
```
